## Task-2 view the distribution of the various features in the data set and calculate their central tendencies

```python
#creating the histogram
plt.figure(figsize=(11,6))
sns.histplot(data=data ,x='fixed acidity',color='green',edgecolor='linen', alpha=0.5,bins=5)
plt.title("histogram of fixed acidity")
plt.xlabel("fixed acidity")
plt.ylabel("count")
plt.show()
```

```python
plt.figure(figsize=(11,6))
sns.histplot(data=data,x='fixed acidity',color='orange',edgecolor='linen',alpha=.5,bins=5)
plt.title("histogram of the fixed acidity")
plt.xlabel("fixed acidity")
plt.ylabel('count')
plt.vlines(data['fixed acidity'].mean(),ymin=0,ymax=4000,color='blue',label='mean')
plt.vlines(data['fixed acidity'].median(),ymin=0,ymax=4000,color='blue',label='median')
plt.legend()
plt.show()
```

Task-3 creating a new pandas series that contains the details of the acid types for a quality

```python
#creating  a new padas called "rep_acid" that contains the details of the representative quality for the different types of acids
rep_acid = pd.Series(
    data = [
        data['fixed acidity'].mean(), 
        data['volatile acidity'].mean(),
        data['citric acid'].mean(),
        data['quality'].mean()
    ],
    index = [
        'fixed acidity',
        'volatile acidity',
        'citric acid',
        'quality'
    ],
    name = 'rep_acid'
)
```
